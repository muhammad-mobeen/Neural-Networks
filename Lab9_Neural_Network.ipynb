```python
import numpy as np
input_data = np.array([-1, 2])
weights = {'node_0': np.array([3, 3]),
           'node_1': np.array([1, 5]),
           'output': np.array([2, -1])}

node_0_input = (input_data * weights['node_0']).sum()
node_0_output = np.tanh(node_0_input)
node_1_input = (input_data * weights['node_1']).sum()
node_1_output = np.tanh(node_1_input)
hidden_layer_outputs = np.array([node_0_output, node_1_output])
output = (hidden_layer_outputs * weights['output']).sum()
print(output)
```

2.   Use ReLU as activation function on the hidden and the output layers

ReLU formula is : f(x) = max(0,x)



```python
import numpy as np

input_data = np.array([3, 5])
weights = {'node_0': np.array([ 2,  4]),
           'node_1': np.array([ 4, -5]),
           'node_2': np.array([-1,  1]),
           'node_3': np.array([ 2,  2]),
           'output': np.array([-3,  7])}

node_0_input = (input_data * weights['node_0']).sum()
node_0_output = np.maximum(0, node_0_input)
node_1_input = (input_data * weights['node_1']).sum()
node_1_output = np.maximum(0, node_1_input)

hidden_layer1_outputs = np.array([node_0_output, node_1_output])

node_2_input = (hidden_layer1_outputs * weights['node_2']).sum()
node_2_output = np.maximum(0, node_2_input)
node_3_input = (hidden_layer1_outputs * weights['node_3']).sum()
node_3_output = np.maximum(0, node_3_input)

hidden_layer2_outputs = np.array([node_2_output, node_3_output])

output = (hidden_layer2_outputs * weights['output']).sum()
print(output)
```
